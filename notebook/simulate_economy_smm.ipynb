```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import pyfixest as pf
 

import os

folder_path = "../output/monte_carlo"
```

```python
df = pd.read_parquet(os.path.join(folder_path,"simulated_panel.parquet"))
df.query('year >= 300',inplace = True)
```

```python
df['Rev'] = np.log(df.E_last_semester)-np.log(df.E_beginning)
df.columns
```

```python
pf.feols("Rev ~ log_sigma  + K + log_D | firm_id",df.query('~Rev.isna()')).summary()
```
